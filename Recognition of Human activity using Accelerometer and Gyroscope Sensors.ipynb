```python
x_train = train_data.drop('Activity',axis=1)
y_train = train_data['Activity']
```

```python
x_test = test_data.drop('Activity',axis=1)
y_test = test_data['Activity']
```

# Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
cli = [10,1,0.5,0.1,0.01,0.003]
```

```python
for j in cli:
    clas = LogisticRegression(C=j)
    clas.fit(x_train,y_train)
    y_pred_train = clas.predict(x_train)
    y_pred_test = clas.predict(x_test)
    trainac = accuracy_score(y_pred_train,y_train)
    print("For value of c =",j,"Training Accuracy score is :",trainac)
    testac = accuracy_score(y_pred_test,y_test)
    print("For value of c =",j,"Testing Accuracy score is :",testac)
```

```python
#confusion matrix(since our test accuracy id more for C=0.5 so taking C=0.5)
clas = LogisticRegression(C=0.5)
clas.fit(x_train,y_train)
y_pred_train = clas.predict(x_train)
y_pred_test = clas.predict(x_test)
print(confusion_matrix(y_pred_train,y_train))
print(confusion_matrix(y_pred_test,y_test))
```

# Now we will apply Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
clf.fit(x_train, y_train)
predsvm = clf.predict(x_train)
a = accuracy_score(y_train,predsvm)
predsvm1 = clf.predict(x_test)
b = accuracy_score(y_test,predsvm1)
print("Trainging Accuracy score:",a)
print("Testing accuracy score:",b)
```

```python
print(confusion_matrix(y_train,predsvm))
print(confusion_matrix(y_test,predsvm1))
```

# Applying Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
#setting random seed
np.random.seed(0)
```

```python
rf_model = RandomForestClassifier(n_jobs=2,random_state=0)
```

```python
rf_model.fit(x_train,y_train)
```

```python
pred_train = rf_model.predict(x_train)
```

```python
pred_test = rf_model.predict(x_test)
```

```python
c = accuracy_score(y_train,pred_train)
d = accuracy_score(y_test,pred_test)
print("Training Accuracy Score:",c)
print("Testing Accuracy Score:",d)
```

```python
print(confusion_matrix(y_train,pred_train))
print(confusion_matrix(y_test,pred_test))
```
